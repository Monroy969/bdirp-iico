```python
class prediccion:
    def inicializacion(self,rutajpg,rutapkl):
        self.pathjpg=rutajpg
        self.pathpkl=rutapkl
        self.alto=28
        self.ancho=28
        self.clases=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

    def load(self):
        self.letra=cv2.imread(self.pathjpg,cv2.IMREAD_GRAYSCALE)

    def conf(self):
        #Definimos las dimensiones para la CNN (numero de imagenes,Alto,Ancho,Numero de canales de color)
        self.IN = np.asarray(self.letra).reshape(-1,self.alto,self.ancho,1)
        self.INF = self.IN.astype('float32')
        
    def evaluacion(self):
        self.y_pred=model.predict(self.INF/255)#Evaluamos los datos
    
    def ordenar(self):
        self.letras=[]
        for n in range(len(self.y_pred)):
            for j in range(len(self.y_pred[n])):
                self.letras.append([self.y_pred[n][j],clases[j]]) #Asignacion de letra a cada exactitud
                
        self.letras.sort(reverse=True)
    
    def guardar(self):
        #Guardado de informacion
        with open(self.pathpkl, "wb") as f:
             pickle.dump(self.letras, f)
```

```python

def calculo_deltas(n_persona,tipo,n_palabra,numero_letras,deep):
    deltas=[]
    datos=[]
    for n_letra in range(1,numero_letras+1):
        o=[]
        with open('C:/Users/'+user+'/Desktop/'+folder+'/Muestras/Persona'+ str(n_persona) +'/'+tipo+'/palabra'+str(n_palabra)+'/letra'+str(n_letra)+'.pkl', "rb") as f: #Abrimos el archivo de predicciones
                obj=pickle.load(f) #obj es un arreglo con dos partes obj[orden de exactitud][letras o valor num√©rico]  

        #Paquete de predicciones
        for x in range(deep):
            o.append(obj[x])
        datos.append(o)

    #Ordenamiento de las deltas de menor a mayor
    deltas=[]
    salida=False
    x=deep-1
    while salida != True:
        for letras in range(len(datos)):
            #print(datos[letras][x-1],datos[letras][x])
            #print(abs(datos[letras][x-1][0]-datos[letras][x][0]),datos[letras][x][1])
            deltas.append(((abs(datos[letras][x-1][0]-datos[letras][x][0]),datos[letras][x][1]),letras))

        salida=True

    deltas.sort()
    return(deltas)
    
def construccion_palabras_deltas(n_persona,tipo,n_palabra,numero_letras,deep):
    estado=False
    deltas=calculo_deltas(n_persona,tipo,n_palabra,numero_letras,deep)
    #print(deltas)
    for posibilidad in range(len(deltas)):

        construccion=""
        i=0
        for n_letra in range(1,numero_letras+1):
            if i==deltas[posibilidad][1]:
                with open('C:/Users/'+user+'/Desktop/'+folder+'/Muestras/Persona'+ str(n_persona) +'/'+tipo+'/palabra'+str(n_palabra)+'/letra'+str(n_letra)+'.pkl', "rb") as f:
                    obj=pickle.load(f) #obj es un arreglo con dos partes 
                                        #obj[orden de exactitud][letras o valor num√©rico]
                
                construccion= construccion+obj[deep-1][1]
            else:
                with open('C:/Users/'+user+'/Desktop/'+folder+'/Muestras/Persona'+ str(n_persona) +'/'+tipo+'/palabra'+str(n_palabra)+'/letra'+str(n_letra)+'.pkl', "rb") as f:
                    obj=pickle.load(f) #obj es un arreglo con dos partes 
                                        #obj[orden de exactitud][letras o valor num√©rico]
                
                construccion= construccion+obj[0][1]
                
            i+=1
        #print(construccion)
        with open('C:/Users/'+user+'/Desktop/'+folder+'/Muestras/Persona'+str(n_persona)+'/palabras.pkl', "rb") as f:
            obj=pickle.load(f)
            
            if obj[n_palabra-1]==construccion:
                print(construccion)
                estado=True
                return estado,construccion
            
    return estado,''
            
def correccion_delta(n_persona,tipo,n_palabra,numero_letras,deep):
    estado=False
    construccion=''
    for x in range(2,deep):
        print(x)
        estado,construccion=construccion_palabras_deltas(n_persona,tipo,n_palabra,numero_letras,x)
        print(estado,construccion)
        if estado==True:
            return estado,construccion
            break
            
    return estado,construccion


    

            

```
