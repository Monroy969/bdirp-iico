```python
import random as rdm #Numeros aleatorios
import os #Creacion de directorios(operating system)
import math #Libreria para usar algunas funciones matematicas
import pickle #manejar archivos y listas
```

```python
#Creacion de directorios
folder='Interfaz grafica pruebas'
user='Monroy969'

#Cargamos diccionarios
diccionarios = [] 
for numero_de_letras in range(2,8):
    diccionario=[]
    n=str(numero_de_letras)
    a=open('C:/Users/'+user+'/Desktop/'+folder+'/Diccionarios/diccionario'+n+'letras.txt', 'r', encoding='utf-8')

    for linea in a:
        diccionario.append(linea)#Se leen las palabras
    a.close

    #Quitamos los escapes de linea
    for x in range(len(diccionario)):
        diccionario[x]=diccionario[x].replace("\n","")

    #Determinamos el tamaño del diccionario 
    lendcn=len(diccionario) #Determinamos el tamaño del diccionario

    diccionarios.append((diccionario,lendcn))
    
```

```python
palabras_prueba=[]
for numero_diccionario in range(len(diccionarios)):
    r=rdm.randint(0,diccionarios[numero_diccionario][1])
    print(r)
    palabras_prueba.append(diccionarios[numero_diccionario][0][r])

print(palabras_prueba)
```
